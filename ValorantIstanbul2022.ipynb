```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('valorant champions istanbul.csv')
df
```

```python
df.info() #Данные в целом ок, смущают лишние символы в столбцах KAST, Prize и распределение мест в Rank 
```

```python
df['KAST'] = df['KAST'].str.replace('%', '') #убираем лишние символы из КАСТ
df['KAST'] = df['KAST'].astype('Float64')
df.head()
```

```python
df['Prize'] = df['Prize'].str.replace('$', '') #Убираем лишние символы из Приза
df['Prize'] = df['Prize'].str.replace('\t', '') 
df['Prize'] = df['Prize'].str.replace(',', '') 
df['Prize'] = df['Prize'].astype('Float64')
df.head()
```

```python
index_56=df.loc[df['Rank'] == '5,6'].index #чиним отображение ранга
index_78=df.loc[df['Rank'] == '7,8'].index

for index in index_56:
    df['Rank'][index]='5'
    
    
for index in index_78:
    df['Rank'][index]='6'

df['Rank'] = df['Rank'].astype('Int64')
```

```python
df.tail() # в командных играх присуждение топ 7-8, или топ 16-20 обычное дело. Присуждение двум командам 5 и 6 места считаю корректным
```

```python
df.info() #С данными закончили - на практике посмотрим, если что то проявится вернемся к вопросу
```

Начнем с тепловой карты и легкого просмотра корреляции

```python
plt.figure(figsize=(10,6))
sns.heatmap(data=df.corr(), annot=True);
```

Тепловая карта позволяет в целом охватить взаимосвязь между данными. Бегло взглянув, мы можем потвердить логичные выводы об обратно-пропорциональной связи достигнутого места по отношению к всем статданным дата фрейма, а так же тот факт, что фраги в играх увеличивают вероятность победить в раунде.  

**Определение лучшего игрока турнира**

```python
# Попробуем визуализировать топ 10 игроков по Киллам, КД, Касту и попаданиям в голову
sns.set_style("darkgrid")
plt.figure(figsize=(10,5))
plt.title("Количество убийств за Турнир")
sns.barplot(x='Player', y='Kill', data=df, order = df.sort_values('Kill').Player.tail(10)) #использован tail потому что head выдает результаты, как будто датафрейм отсортирован от меньшего к большему
```

```python
sns.set_style("darkgrid")
plt.figure(figsize=(10,5))
plt.title("Most Valuable Player Турнира")
sns.barplot(x='Player', y='K/D', data=df, order = df.sort_values('K/D').Player.tail(10))
```

```python
sns.set_style("darkgrid")
plt.figure(figsize=(10,5))
plt.title("Самый точный игрок Турнира")
sns.barplot(x='Player', y='HS %', data=df, order = df.sort_values('HS %').Player.tail(10))
```

```python
sns.set_style("darkgrid")
plt.figure(figsize=(10,5))
plt.title("Эффективность игрока в общем плане")
sns.barplot(x='Player', y='KAST', data=df, order = df.sort_values('KAST').Player.tail(10))
```

По текущим данным мы можем сделал вывод, что игрок **yay** стал самым смертоносным игроком турнира, а по проценту Хедшотов держится в 5 пятерке. КАСТ (Kill, Assist, Survive & Traded) больше нужен для определенния эффективности игроков, которые играют на позиции суппортов, об этом ниже.

Есть такие игроки, как **kNigg**, которые статистически показывают хорошие результаты, однако высокий показатель КД обусловлен их вылетом в начальной стадии турнира и небольшом количестве фрагов по сравнению с тем же **yay**, поэтому конкурент из **kNigg** слабый. На втором месте выделю **pANcada** - на роли опорного игрока у него очень высокий процент попаданий в голову и первое место по КАСТ, что и является его главным критерием оценки.

**Посмотрим на команды**

```python
plt.figure(figsize=(12,6)) # оценим общие фраги на команду
sns.boxplot( x=df["Team"], y=df["Kill"])
```

Больше всего внимания заслуживают наши финалисты - **LOUD** и **OPTC**. 

**LOUD** - на турнире в стамбуле им не было равных - на групповом этапе турнира и в Плей-офф они закрывали другие команды с разгромом - 2:0, кроме **OPTC** - они в финале сыграли 3:1 в пользу **LOUD**. Этим обусловленно небольшое количество фрагов по сравнению с долгоиграющими свои серии **OPTC** и **DRX**. Также можно отметить отсутсвие яркого индивидуального игрока - черта присущая слабым коллективам.  

**OPTC** - в данном коллективе выброс - это **yay**. Очень яркий индивидуальный исполнитель, который выделяется на фоне остальных - он настрелял фрагов на ~ 100 больше чем его тиммейты, возможно, благодаря его заслугам этот коллектив и смог прорваться ко второму месту с боем. Разница во фрагах с **LOUD** обусловлена тем, что **OPTC** в борьбе вырывали победу (очень много 2:1), а в финале верхней сетки упали в нижнюю сетку, чем сыграли на одну финальную игру больше, которая так же была длинной и закончилась 3:2 в пользу **OPTC**. 

```python
sns.set_style("darkgrid")
plt.figure(figsize=(10,6))
plt.title("Распределение ролей в команде",fontsize=10)
sns.countplot(x='Role',hue='Team',data=df)
plt.show()
```

Как мы видим, победители турнира **LOUD** имеют классический вид распределения ролей, без Flex - гибкого игрока, который может (условно) сыграть на любой позиции. Больше всего меня привлекла команда **XSET**, которая имеет 5 flex игроков.
